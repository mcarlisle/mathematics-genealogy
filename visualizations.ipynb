```python
import pickle
import plotly.express as px
```

```python
with open('../count_rf_20190729.pickle', 'rb') as f:
    classifier = pickle.load(f)
```

```python
# MSC per year
with open('msc_per_year_df.pickle', 'rb') as f:
    msc_per_year_df = pickle.load(f)
```

```python
fig = px.bar(msc_per_year_df, x="msc", y="count", text="msc_name", hover_name="msc_name", 
             color="msc", animation_frame="year", animation_group="year", range_y=[0,1300],
             title="MSC breakdown per year, all countries")
fig.show()
```
